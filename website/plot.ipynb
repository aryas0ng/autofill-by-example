```python
import matplotlib.pyplot as plt
import time

from local_main import autofill, load_csv, check
```

```python
number_to_string = {
        1: 'one',
        2: 'two',
        3: 'three',
        4: 'four',
        5: 'five',
        6: 'six',
        7: 'seven',
        8: 'eight',
        9: 'nine',
        10: 'ten'
    }
```

```python
# Record and plot the execution time of datasets under the weighted sum formula
# with different number of columns 
def weighted_sum_plot():
    times = []
    for i in range(2,11):
        n = number_to_string.get(i)
        file = "../tests/weighted_sum_of_" + n + "_cols.csv"
        dataset = load_csv(file)

        start_time = time.time()
        result = autofill(dataset)
        end_time = time.time()

        duration = (end_time - start_time) * 1000

        times.append(duration)

    plt.plot(list(range(2, 11)), times)

    plt.xlabel('Number of Columns')
    plt.ylabel('Execution Time (ms)')
    plt.title('Weighted Sum Execution Time Plot')

    plt.show()
```

```python
weighted_sum_plot()
```

```python
# Record and plot the execution time of datasets under the weighted sum formula
# with different number of columns 
def multiple_rows_plot():
    times = []
    for i in range(1,11):
        n = number_to_string.get(i)
        file = "../tests/sum_of_" + n + "_rows.csv"
        dataset = load_csv(file)

        start_time = time.time()
        _, result = autofill(dataset)
        end_time = time.time()

        duration = (end_time - start_time) * 1000

        times.append(duration)

        file = file[:-4] + "_expected.csv"
        check(result, file)

    plt.plot(list(range(1, 11)), times)

    plt.ylim(0, 50)

    plt.xlabel('Number of Rows')
    plt.ylabel('Execution Time (ms)')
    plt.title('Multiple Rows Execution Time Plot')

    plt.show()
```

```python
multiple_rows_plot()
```
